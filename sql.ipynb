```python
import pandas
import csv, sqlite3
```

```python
conn= sqlite3.connect("drugbank.db")
```

```python
df = pandas.read_csv('WorldHappiness_Corruption.csv')
df.to_sql('drugbank', conn, if_exists='append', index=False) 
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("select * from drugbank")
col_name_list = [tuple[0] for tuple in cursor.description]
print (col_name_list)
```

```python
# 执行查询语句:
cursor.execute('select Country from drugbank ')
# 获得查询结果集:
values = cursor.fetchall()
# 遍历打印输出
for smile in values:
    print(smile)
```
