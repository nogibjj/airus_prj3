```python
#list country in order of happiness_score
connection = sqlite3.connect('happy.db')
cursor = connection.cursor()

sql_1 = """
SELECT Country
FROM happy
ORDER BY happiness_score;
"""

cursor.execute(sql_1)
allteams = cursor.fetchall() # a list of tuples
print(allteams)
```

```python
#find contry group by continent and order by health in descending
connection = sqlite3.connect('happy.db')
cursor = connection.cursor()

sql_2 = """
SELECT country
FROM happy
GROUP BY continent
ORDER BY health DESC;
"""

cursor.execute(sql_2)
allteams = cursor.fetchall() # a list of tuples
print(allteams)
```

```python
#show country and happiness_score in North America in order of freedom
connection = sqlite3.connect('happy.db')
cursor = connection.cursor()

sql_3 = """
SELECT distinct country, happiness_score
FROM happy
where continent = "North America"
ORDER BY freedom
"""

cursor.execute(sql_3)
playersBOS = cursor.fetchall() # a list of tuples
print(playersBOS)
```

```python
#find country gdp above 1.6
connection = sqlite3.connect('happy.db')
cursor = connection.cursor()

sql_3 = """
SELECT  country
FROM happy
where gdp_per_capita>1.6;
"""

cursor.execute(sql_3)
mvpplayers = cursor.fetchall() # a list of tuples
df2 = pd.DataFrame(mvpplayers)
print(df2)
```
